```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_df, palette='viridis')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='species', y='sepal_length', data=iris_df, palette='Set2')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal_width', data=iris_df, palette='pastel')
plt.title('Bar Plot of Average Sepal Width by Species')
plt.show()
```
